```python
import json
```

```python
# Below text file is on home directory of EMR master node.
with open("meetup_test") as f:
    f = f.readline().split("|||")[0]
    record = json.loads(f)
```

```python
# meetings
for result in record['results']:
    print(result['id'], result['name'], result['city'], result['members'])
```

```python
# organizer
for result in record['results']:
    organizer = result.get('organizer')
    if organizer is None:
        continue
    print(result['id'], organizer['member_id'], organizer['name'])
```

```python
# category
for result in record['results']:
    category = result.get('category')
    if category is None:
        continue
    print(result['id'], category['id'], category['name'])
```

```python
# topics
tdf = []
for result in record['results']:
    topics = result.get('topics')
    if topics is None:
        continue
    # meeting id, topic id, names
    for topic in topics:
        print(result['id'], topic['id'], topic['name'])
        tdf.append([result['id'], topic['id'], topic['name']])
```
